## Funciones

```python
from sklearn.metrics import recall_score, f1_score, roc_auc_score, cohen_kappa_score, balanced_accuracy_score

def calculate_metrics(classifier_name, y_test, y_pred):
    """
    Calcula las métricas de rendimiento para un clasificador dado.

    Parameters:
    - classifier_name: str, nombre del clasificador ('KNN', 'LR', 'MLP')
    - y_test: array-like, etiquetas verdaderas
    - y_pred: array-like, etiquetas predichas

    Returns:
    - metrics: dict, diccionario con métricas calculadas
    """
    # Crear un diccionario para almacenar las métricas
    metrics = {}

    # Calcular las métricas
    metrics[f'recall_1_{classifier_name.lower()}'] = recall_score(y_test, y_pred, pos_label=1)
    metrics[f'recall_0_{classifier_name.lower()}'] = recall_score(y_test, y_pred, pos_label=0)
    metrics[f'f1_1_{classifier_name.lower()}'] = f1_score(y_test, y_pred, pos_label=1)
    metrics[f'f1_0_{classifier_name.lower()}'] = f1_score(y_test, y_pred, pos_label=0)
    metrics[f'roc_auc_{classifier_name.lower()}'] = roc_auc_score(y_test, y_pred)
    metrics[f'cohen_kappa_{classifier_name.lower()}'] = cohen_kappa_score(y_test, y_pred)
    metrics[f'balanced_accuracy_{classifier_name.lower()}'] = balanced_accuracy_score(y_test, y_pred)

    return metrics


def print_metrics(metrics):
    """
    Imprime las métricas de manera estructurada.

    Parameters:
    - metrics: dict, diccionario con métricas calculadas
    """
    # Convertir el diccionario en un DataFrame
    metrics_df = pd.DataFrame.from_dict(metrics, orient='index', columns=['Valor'])

    # Reiniciar el índice para mejor visualización
    metrics_df.reset_index(inplace=True)

    # Cambiar nombres de las columnas para mayor claridad
    metrics_df.columns = ['Métrica', 'Valor']

    # Imprimir el DataFrame de manera bonita
    print(metrics_df.to_string(index=False))


def save_metrics_to_csv(metrics, filename='metrics.csv', append=True):
    """
    Guarda las métricas en un archivo CSV. Si el archivo ya existe, se añade al archivo existente.

    Parameters:
    - metrics: dict, diccionario con métricas calculadas
    - filename: str, nombre del archivo CSV (por defecto 'metrics.csv')
    - append: bool, si es True, agrega al archivo existente
    """
    # Convertir el diccionario en un DataFrame
    metrics_df = pd.DataFrame.from_dict(metrics, orient='index', columns=['Valor'])
    metrics_df.reset_index(inplace=True)
    metrics_df.columns = ['Métrica', 'Valor']

    # Guardar el DataFrame en un archivo CSV
    if append:
        # Agregar al archivo existente
        metrics_df.to_csv(filename, mode='a', header=False, index=False)
    else:
        # Sobrescribir el archivo existente
        metrics_df.to_csv(filename, mode='w', index=False)

    print(f'Métricas guardadas en {filename}')





```

```python
from sklearn.model_selection import train_test_split

def procedimiento(X_train_res, y_train_res, param_grid_KNN=param_grid_KNN, param_grid_LR=param_grid_LR, param_grid_MLP=param_grid_MLP):
    # División de datos
    X_train, X_test_val, y_train, y_test_val = train_test_split(X_train_res, y_train_res, train_size=0.8, random_state=42)
    X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5, random_state=42)

    # Entrenar y evaluar modelos
    best_knn_model, knn_metrics, knn_params = train_and_evaluate(
        'KNN', KNeighborsClassifier(), param_grid_KNN, X_train, y_train, X_val, y_val, X_test, y_test
    )

    best_lr_model, lr_metrics, lr_params = train_and_evaluate(
        'LR', LogisticRegression(), param_grid_LR, X_train, y_train, X_val, y_val, X_test, y_test
    )

    best_mlp_model, mlp_metrics, mlp_params = train_and_evaluate(
        'MLP', MLPClassifier(random_state=42), param_grid_MLP, X_train, y_train, X_val, y_val, X_test, y_test
    )

    # Diccionarios de resultados
    modelos = {'KNN': best_knn_model, 'LR': best_lr_model, 'MLP': best_mlp_model}
    metricas = {'KNN': knn_metrics, 'LR': lr_metrics, 'MLP': mlp_metrics}
    mejores_parametros = {'KNN': knn_params, 'LR': lr_params, 'MLP': mlp_params}

    # Retornar modelos, métricas, parámetros y conjuntos de datos
    return modelos, metricas, mejores_parametros, X_train, y_train, X_val, y_val, X_test, y_test


def train_and_evaluate(classifier_name, model, param_grid, X_train, y_train, X_val, y_val, X_test, y_test):
    # Configuración de búsqueda en cuadrícula
    grid_search = GridSearchCV(model, param_grid, scoring='f1', cv=5)
    grid_search.fit(X_val, y_val)

    # Mejor modelo y parámetros
    best_params = grid_search.best_params_
    best_model = model.set_params(**best_params)
    best_model.fit(X_train, y_train)

    print(best_params)

    # Evaluación
    y_pred = best_model.predict(X_test)
    metrics = calculate_metrics(classifier_name, y_test, y_pred)
    print_metrics(metrics)

    save_metrics_to_csv(metrics)

    return best_model, metrics, best_params

```

## Funciones escribir en csv

```python
import os

import pandas as pd

import pandas as pd

import pandas as pd

def guardar_metricas_comparadas(metrics_smote, metrics_ros, filepath, descripcion):
    # Crear una lista para almacenar los resultados
    resultados = {'Métrica': []}

    # Definir los nombres de los modelos
    modelos = metrics_smote.keys()

    # Agregar columnas para cada modelo con SMOTE y ROS
    for modelo in modelos:
        resultados[f'{modelo}_SMOTE'] = []
        resultados[f'{modelo}_ROS'] = []

    # Definir las métricas a incluir
    metricas = ['recall_1', 'recall_0', 'f1_1', 'f1_0', 'roc_auc', 'cohens_kappa', 'balanced_accuracy']

    # Agregar valores a resultados para cada métrica
    for metric in metricas:
        resultados['Métrica'].append(metric)
        for modelo in modelos:
            # Agregar los valores de SMOTE y ROS para cada métrica y modelo
            resultados[f'{modelo}_SMOTE'].append(metrics_smote[modelo].get(f'{metric}_{modelo.lower()}'))
            resultados[f'{modelo}_ROS'].append(metrics_ros[modelo].get(f'{metric}_{modelo.lower()}'))

    # Convertir los resultados a un DataFrame
    df_metrics = pd.DataFrame(resultados)

    # Escribir la descripción y la tabla en el archivo CSV
    with open(filepath, 'a') as file:
        # Escribir la descripción como primer línea
        file.write(f"{descripcion}\n\n")

    # Guardar el DataFrame debajo de la descripción
    df_metrics.to_csv(filepath, mode='a', index=False)


import pandas as pd

def guardar_mejores_parametros_comparados(params_smote, params_ros, filepath, descripcion):
    # Inicializar un diccionario para almacenar los resultados
    parametros_data = {('Parámetro', ''): []}

    # Añadir claves dinámicamente para cada técnica y modelo
    for metodo in ['SMOTE', 'Random Over Sampling']:
        for modelo in ['KNN', 'LR', 'MLP']:
            parametros_data[(metodo, modelo)] = []

    # Determinar los parámetros únicos de todos los modelos en SMOTE y ROS
    all_params = set()
    for modelo_params in [params_smote, params_ros]:
        for modelo in modelo_params:
            all_params.update(modelo_params[modelo].keys())

    # Rellenar el diccionario con los parámetros
    for param in all_params:
        parametros_data[('Parámetro', '')].append(param)
        # Extraer y agregar parámetros para cada modelo y técnica
        for modelo in ['KNN', 'LR', 'MLP']:
            # Obtener el valor de SMOTE o None si no existe el parámetro
            smote_value = params_smote.get(modelo, {}).get(param, None)
            parametros_data[('SMOTE', modelo)].append(smote_value)

            # Obtener el valor de ROS o None si no existe el parámetro
            ros_value = params_ros.get(modelo, {}).get(param, None)
            parametros_data[('Random Over Sampling', modelo)].append(ros_value)

    # Escribir la descripción y la tabla en el archivo CSV
    with open(filepath, 'a') as file:
        # Escribir la descripción como primer línea
        file.write(f"{descripcion}\n\n")

    # Convertir a DataFrame y guardar en CSV
    df_parametros = pd.DataFrame(parametros_data)
    df_parametros.to_csv(filepath,mode='a' ,index=False)

```

```python


# Definir el modelo en TensorFlow
model = models.Sequential()

# Capa de entrada
model.add(layers.InputLayer(input_shape=(X_train3s.shape[1],)))

# Capa oculta con activación tanh y regularización L2
model.add(layers.Dense(100, activation='tanh', 
                       kernel_regularizer=regularizers.l2(0.001)))

# Capa de salida (ajustar el tamaño y la activación según el problema)
model.add(layers.Dense(1, activation='sigmoid'))  # Para clasificación binaria

# Compilar el modelo (solo para inicializarlo)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Resumen del modelo
model.summary()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim


input_size = input_shape  # Debes reemplazar esto con el tamaño de tus características de entrada
output_size = 1  # Dependiendo del problema (1 si es regresión, o número de clases si es clasificación)

sklearn_weights = weights
sklearn_biases = intercepts 



class MyModel(nn.Sequential):
    # Otras funciones de la clase
    
    def shape_input(self, x, shape, channel_first=True):
        # Verifica que 'shape' sea una tupla o lista de enteros.
        if not isinstance(shape, (tuple, list)):
            raise ValueError("El argumento 'shape' debe ser una tupla o lista de enteros.")
        
        # Asegúrate de que la longitud del shape coincida con el número de elementos en x
        try:
            return x.view(*shape)  # Desempaqueta la tupla de 'shape'
        except Exception as e:
            print(f"Error al cambiar la forma del tensor: {e}")
            raise
    


# Crear el modelo usando nn.Sequential
model_pt = MyModel(
    nn.Linear(input_size, 100),  # Capa oculta de 100 neuronas
    nn.Tanh(),                   # Función de activación Tanh
    nn.Linear(100, output_size),  # Capa de salida (1 neurona)
    nn.Sigmoid()                 # Activación sigmoide para clasificación binaria
)

# Transferir los pesos y sesgos a PyTorch
with torch.no_grad():
    # Para la primera capa (capa oculta)
    model_pt[0].weight = nn.Parameter(torch.tensor(sklearn_weights[0].T, dtype=torch.float32))  # Capa oculta
    model_pt[0].bias = nn.Parameter(torch.tensor(sklearn_biases[0], dtype=torch.float32))        # Bias de la capa oculta
    
    # Para la capa de salida
    model_pt[2].weight = nn.Parameter(torch.tensor(sklearn_weights[1].T, dtype=torch.float32))  # Capa de salida
    model_pt[2].bias = nn.Parameter(torch.tensor(sklearn_biases[1], dtype=torch.float32))        # Bias de la capa de salida

# Optimizador Adam como en tu modelo de sklearn
optimizer = optim.Adam(model_pt.parameters(), lr=0.001)

# Preparar los datos para la predicción
# x_test debe ser un tensor de PyTorch
# Por ejemplo, si tienes datos en un array de NumPy:

# Supongamos que tus datos de prueba están en un numpy array llamado X_test
X_test_tensor = torch.tensor(X_test3s.values, dtype=torch.float32)

# Realizar predicciones
model_pt.eval()  # Cambia el modelo a modo de evaluación
with torch.no_grad():
    y_pred_tensor = model_pt(X_test_tensor)  # Realiza la predicción
    
    y_pred = (y_pred_tensor > 0.5).int()  # Para clasificación binaria



# Si necesitas las predicciones como un array de NumPy
y_pred_pt = y_pred.numpy()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_feature_importances(feature_matrix, importances):
    """
    Función para graficar la importancia de las características con colores diferenciados
    para contribuciones positivas (naranja) y negativas (azul). 
    
    Parámetros:
    - feature_matrix: DataFrame de pandas con las características.
    - importances: array de numpy o lista con las importancias de las características.
    
    Salida:
    - Gráfico de barras con las importancias de las características organizadas por 
      el valor absoluto de las atribuciones.
    """
    # Asegurarse de que las importancias son un array de numpy
    importances = np.array(importances).flatten()  # Aplanar en caso de que sea multidimensional

    # Obtener los nombres de las características
    feature_names = feature_matrix.columns.tolist()

    # Crear un DataFrame para las importancias
    attrib_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importances
    })

    # Ordenar el DataFrame por el valor absoluto de las importancias
    attrib_df['AbsImportance'] = attrib_df['Importance'].abs()
    attrib_df = attrib_df.sort_values(by='AbsImportance', ascending=False)

    # Colores para las barras: naranja para relevancias positivas, azul para negativas
    colors = attrib_df['Importance'].apply(lambda x: 'orange' if x > 0 else 'blue')

    # Graficar las importancias
    plt.figure(figsize=(10, 6))
    plt.barh(attrib_df['Feature'], attrib_df['Importance'], color=colors)
    plt.xlabel('Relevancia')
    plt.title('Importancia de las Características (Colores: Naranja=Positiva, Azul=Negativa)')
    plt.axvline(0, color='red', linestyle='--')  # Línea vertical en 0
    plt.gca().invert_yaxis()  # Invertir eje Y para que las barras más grandes estén arriba
    plt.show()

# Ejemplo de uso
feature_matrix = pd.DataFrame(X_test3s.values, columns=X_test3s.columns)
importances = attributions_np  # O cualquier otro array con las importancias
plot_feature_importances(feature_matrix, importances)
```

```python

def modify_instance_based_on_lime(instance, lime_exp, num_perturbations=10):
    perturbed_instances = []
    for _ in range(num_perturbations):
        perturbed_instance = instance.copy()
        perturbed_instances.append(perturbed_instance)
    return np.array(perturbed_instances)

def calculate_lime_fidelity(explainer, model_predict_proba, X_test, num_samples=100):
    fidelity_scores = []
    
    for i in range(num_samples):
        test_instance = X_test[i]
        predicted_label = np.argmax(model_predict_proba(test_instance.reshape(1, -1))[0])

        exp = explainer.explain_instance(
            data_row=test_instance, 
            predict_fn=model_predict_proba,
            labels=[predicted_label],  
            num_features=len(feature_names)
        )
        modified_instances = modify_instance_based_on_lime(test_instance, exp.as_list(label=predicted_label))
        
        if modified_instances is not None:
            original_prediction = model_predict_proba(test_instance.reshape(1, -1))[0]
            perturbed_predictions = np.array([model_predict_proba(instance.reshape(1, -1))[0] for instance in modified_instances])
        
        fidelity = np.mean(np.argmax(perturbed_predictions, axis=1) == predicted_label)
        fidelity_scores.append(fidelity)
    
    return np.mean(fidelity_scores)




```

```python
import numpy as np

def sum_n_val(index, analysis):
    """ Suma los valores de los índices relevantes para la clasificación utilizando NumPy. """
    # Convertir a arrays de NumPy si no lo son
    analysis = np.asarray(analysis)
    index = np.asarray(index)
    # Seleccionar los valores directamente en lugar de usar bucles
    selected_values = analysis[:, index]
    # Sumar a lo largo de la segunda dimensión (por los índices)
    return selected_values.sum(axis=1)

def flatten_3d_array(array):
    """ Aplica el aplanamiento de relevancias a un array 3D. """
    flatt_arr = []
    for analysis_select in array:
        # sum_prom_analysis = np.mean(analysis_select, axis=2).flatten() # dim 784 PROMEDIA 3canales
        sum_prom_analysis = analysis_select.reshape(-1) # dim 2352 Aplana un tensor 3D concatenando los canales uno tras otro
        # print(f"SHAPE FLATTEN ANALYSISS: {np.asarray(sum_prom_analysis).shape}")
        flatt_arr.append(np.asarray(sum_prom_analysis))
    return flatt_arr

def lse_calculate(relevances, class_index, filter=0):
    """
    Calcula el IFI usando el array de relevancias (ex. LIME) ya generado.

    Parámetros:
    - relevances: Array de relevancias (ex. generado por LIME) que contiene la relevancia de las neuronas de salida.
    - predict: Predicción del modelo (salida de la red para la entrada dada).
    - class_index: Índice de la neurona que tiene la clasificación más probable.
    - filter: Umbral de filtro para los valores de relevancia.

    Retorna:
    - ifi: Valor IFI calculado.
    - relevances: Resultado del análisis.
    """

    print(f"IFI_Calculate; Original SHAPE: {np.array(relevances).shape}")

    # Aplanar los vectores si es necesario
    if relevances[0].ndim == 3:
        print("DIMENSION X,Y,Z")
        f_analysis = flatten_3d_array(relevances)
        print("Len flatten analysis", len(f_analysis))
    else:
        print(f"DIMENSION X,Y, Original SHAPE: {np.array(relevances).shape}")
        f_analysis = [arr.flatten() for arr in relevances]
        print(f"SHAPE FLATTEN ARRAYS {np.array(f_analysis).shape}")
    
    # Obtener el vector de relevancia aplanado para la neurona de la clase predicha
    f_analysis_class = np.asarray(f_analysis[class_index])

    # Verificar si los valores están fuera del rango [0,1]
    max_val = np.max(f_analysis_class)
    min_val = np.min(f_analysis_class)

    # Normalizar solo si es necesario
    if max_val > 1 or min_val < 0:
        print("Normalizando porque los valores están fuera del rango [0,1]")
        
        # Normalización manual al rango [0,1]
        norm_class = (f_analysis_class - min_val) / (max_val - min_val)
        f_analysis_normalized = (f_analysis - np.min(f_analysis)) / (np.max(f_analysis) - np.min(f_analysis))
    else:
        print("No es necesario normalizar, los valores están dentro del rango [0,1]")
        norm_class = f_analysis_class
        f_analysis_normalized = f_analysis
    
    print(f"MAX NORMA: {np.max(norm_class)}, MIN NORMA: {np.min(norm_class)}")

    # Usar np.where para encontrar los índices más relevantes por encima del filtro
    index_max = np.where(norm_class >= filter)[0]

    # Imprimir para verificar los índices encontrados
    print(f"Índices con valores >= {filter}: {index_max}")
    print(f"Cantidad de índices encontrados: {len(index_max)}")

    # Sumar los n-valores de mayor relevancia para la clasificación (por cada neurona de salida)
    sum_vals = sum_n_val(index=index_max, analysis=f_analysis_normalized)

    # Calcular el valor de relevancia de la neurona de clasificación frente al resto
    val_class = sum_vals[class_index]
    val_other = sum(sum_vals[i] for i in range(len(sum_vals)) if i != class_index)

    print(f"SUM CLASS: {val_class}\nSUM OTHER: {val_other}")

    # Calcular el IFI
    if val_other > 0:
        ifi = (len(relevances) * val_class) / val_other
    elif val_other == 0:
        ifi = 0
    else:
        ifi = float('inf')

    return ifi
```

```python
# Funcion para preprocesar la salida de lime
def procesar_salida_lime_optimizado(salida_lime, num_caracteristicas=6):
    # Crear dos arreglos de ceros del tamaño del número de características
    relevancias_clase_0 = [0] * num_caracteristicas
    relevancias_clase_1 = [0] * num_caracteristicas
    
    # Recorrer las relevancias de ambas clases simultáneamente
    for (indice_0, relevancia_0), (indice_1, relevancia_1) in zip(salida_lime[0], salida_lime[1]):
        relevancias_clase_0[indice_0] = relevancia_0
        relevancias_clase_1[indice_1] = relevancia_1
    
    # Devolver las relevancias para ambas clases
    return [relevancias_clase_0, relevancias_clase_1]




# Llamar a la función
resultado = procesar_salida_lime_optimizado(exp)

# Mostrar el resultado
print(resultado)
```

# Integrated methods

```python
def calculate_method_scores(lse_values, method_names):
    """
    Puntúa los métodos de acuerdo a su IFI (mientras más cerca de 1, mejor).
    La suma de las puntuaciones debe ser igual a 1.

    Args:
        lse_values (list): Lista de valores IFI para cada método.
        method_names (list): Lista de nombres de los métodos.

    Returns:
        dict: Diccionario con la puntuación de cada método.
    """
    if len(lse_values) != len(method_names):
        raise ValueError("El número de valores LSE debe coincidir con el número de métodos.")

    # Calcular la cercanía a 1 usando el inverso de la distancia absoluta de 1
    ifi_closeness = [1 / (1 + abs(1 - lse)) for lse in lse_values]

    # Normalizar las puntuaciones para que la suma sea 1
    total_closeness = sum(ifi_closeness)
    method_scores = [score / total_closeness for score in ifi_closeness]

    # Crear un diccionario de resultados
    method_scores_dict = {method_names[i]: method_scores[i] for i in range(len(method_scores))}

    return method_scores_dict
```

```python
scores = [1.3, 1.5, 0.99, 4]
methods = ["LIME", "SHAP", "Silency maps", "Integrated Gradients"]

valores = calculate_method_scores(scores, methods)
print(valores)
```

```python
# Función para calcular los pesos para cada imagen
def calculate_weights_for_image(lse_values):
    """
    Calcula los pesos para cada método basado en los LSEs de una imagen.
    Mientras más cercano el LSE a 1, mayor será el peso del método.

    Args:
        lse_values (list): Lista de valores de LSE para cada método en una imagen.

    Returns:
        list: Pesos normalizados para aplicar en el operador OWA.
    """
    # Calcular la cercanía a 1 usando el inverso de la distancia absoluta de 1
    ifi_closeness = [1 / (1 + abs(1 - lse)) for lse in lse_values]

    # Normalizar las puntuaciones para que la suma sea 1
    total_closeness = sum(ifi_closeness)
    weights = [score / total_closeness for score in ifi_closeness]

    return weights
```

```python
scores = [1.3, 0.7, 0.99, 4.0]
methods = ["LIME", "SHAP", "Silency maps", "Integrated Gradients"]

valores = calculate_weights_for_image(scores)
print(valores)
```
